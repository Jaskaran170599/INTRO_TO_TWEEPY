### Importing necessary Packages

```python
from tweepy.streaming import StreamListener
from tweepy import OAuthHandler
from tweepy import Cursor
from tweepy import Stream
from tweepy import API
import matplotlib.pyplot as plt
import pandas as pd
import jsonlines
```

### Key Variables

```python
#Just commited
# enter the keys
access_token=""
access_token_secret=""
api_secret_key=""
api_key=""
```

### Authentication

```python
auth = OAuthHandler(api_key,api_secret_key)
auth.set_access_token(access_token, access_token_secret)
auth_api = API(auth)
```

### Getting User 

```python
item = auth_api.get_user("@truTVjokers")
```

### Collecting User's tweets according to the tweet number

```python
tweets=[]
j=item.statuses_count
for status in Cursor(auth_api.user_timeline, id="midasIIITD").items():
    status._json["tweetnum"]=j   #just to record the tweet number otherwise we could simply compare tweetid
    tweets.append(status._json)
    j-=1
```

### Dumping into a JSONLines FILE 

```python
with jsonlines.open('tweets.jsonl', 'w') as writer:
    writer.write_all(tweets)
```

### Loading the tweets.jsonl file

```python
midas_tweets=jsonlines.Reader(open("tweets.jsonl","r",encoding="UTF-8"))  #Iterator
```

### Info about the Tweets

● The text of the tweet.
● Date and time of the tweet.
● The number of favorites/likes.
● The number of retweets.
● Number of Images present in Tweet. If no image returns None

### Data Processing

```python
cols=["Tweet number:","Date and Time","Text","num of Fav/Likes","num of Retweets","num of Images present"] #column name

data=[]

for i in midas_tweets:
    j=0
    if "media" not in i["entities"].keys(): #if no media file
        j=None
    else:
        for k in i["entities"]["media"]:    #iterating through every media file
            if k["type"]=="photo":
                j+=1
    data.append([i["tweetnum"],i["created_at"],i["text"],i["favorite_count"],
                i["retweet_count"],j])
```

### DataFraming

```python
data=pd.DataFrame(data)
data.columns=cols
data
```
